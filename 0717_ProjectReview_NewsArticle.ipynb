```
dat = pd.read_csv('민주vs미래.csv', encoding='cp949')
dat['total'] = dat.민주 + dat.미래
```

```
fig, ax = plt.subplots(figsize=(12,6))
grp = sns.scatterplot(x='민주', y='미래', hue='신문사', alpha=.7, data=dat, size="total", sizes = (0,5000), legend = 'brief')
plt.xlim(0,350)
plt.ylim(0,400)
plt.legend(loc='upper left', bbox_to_anchor=(-0.05, -0.2),fancybox=True, shadow=True, ncol=5)
for i in range(20):
  plt.text(dat.민주[i], dat.미래[i], dat.신문사[i])

h,l = grp.get_legend_handles_labels()

#h[:30], l[:30]

col_lgd = plt.legend(h[:20], l[:30] ,loc='upper left', 
                    bbox_to_anchor=(-0.05, -0.1), fancybox=True, shadow=True, ncol=5)
```
