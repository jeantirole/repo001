```python
for i in range(1,6,1): 
    score = input("당신의 점수를 입력해주세요: ")
    score = int(score)

    if(score >= 90.0):
        print("학점은 A입니다.")
    elif(score >=80):
        print("학점은 B입니다.")
    elif(score >=70):
        print("학점은 C입니다.")
    elif(score >=60):
        print("학점은 D입니다.")
    else:
        print("학점은 F입니다.")
```
